#### Import required libs

```python
import sys
sys.path.append('/home/jupyter/app')

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from src.io_utils import load_dataframe
```

#### Define constants

```python
GS_DIR_FEATURES = "gs://pcqm4mv2/data/features"
GS_DIR_AUTOML = "gs://pcqm4mv2/data/automl"
```

#### Load datasets

```python
df_train = load_dataframe("train.csv", GS_DIR_FEATURES)

print("df_train shape:", df_train.shape)
df_train.head()
```

```python
df_val = load_dataframe("valid.csv", GS_DIR_FEATURES)

print("df_val shape:", df_val.shape)
df_val.head()
```

#### Split `df_train` into `train` and `validate`

```python
train_idx, validate_idx = train_test_split(df_train.index.values, test_size=0.025, random_state=2912)
```

```python
train_idx
```

```python
validate_idx
```

#### Use `df_val` as `test`

```python
test_idx = df_val.index.values
test_idx
```

#### Concatenate `df_train` and `df_val` to a unique dataframe

```python
df_unique = pd.concat([df_train, df_val])
df_unique
```

```python
assert df_unique.shape[0] == len(df_train) + len(df_val)
assert df_unique.shape[1] == df_train.shape[1]
assert df_unique.shape[1] == df_val.shape[1]
```

#### Create `splitcol`

```python
df_unique.loc[train_idx.tolist(), "splitcol"] = "TRAIN"
df_unique.loc[validate_idx.tolist(), "splitcol"] = "VALIDATE"
df_unique.loc[test_idx.tolist(), "splitcol"] = "TEST"
```

```python
df_unique["splitcol"].value_counts(normalize=True)
```

#### Save dataset to GCS

```python
df_unique.to_csv(f"{GS_DIR_AUTOML}/train_val_test.csv")
```
